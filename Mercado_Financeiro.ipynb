```python
#Uma representação possivel dos dados deverá ser atraves das series temporais
```

```python
bd_dados = dados.set_index('Data')
```

```python
bd_dados.head()
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16, 5))
plt.title("Analise das acções financeiras da Magalu - Fechamento", fontsize=15, loc="left")
plt.plot(bd_dados.index, bd_dados['Fechamento'] )

plt.xlabel('Periodo da Coração')
plt.ylabel('Valor da Ação (R$)');
```

```python
#Plotando outra serie temporal no mesmo gráfico para analisar o fechamento de anual, trinta dias e cinco semanas.
```

```python
bd_dados.tail()
```

```python
Media_Movel = bd_dados['Fechamento'].rolling(5).mean()
Media_Tendencia = bd_dados['Fechamento'].rolling(30).mean()

plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16, 5) )
plt.title('Análise das ações da magalu - Fechamento', fontsize=15, loc='left')

plt.plot( bd_dados.index, bd_dados['Fechamento'] )
plt.plot( bd_dados.index, Media_Movel )
plt.plot( bd_dados.index, Media_Tendencia )

plt.xlabel('Período da Coração')
plt.ylabel('Valor da Ação (R$)');
```

```python
#Analisar um boxplot Geral para observar a variavel de Fachamento
```

```python
plt.figure(figsize=(7, 4))
sns.boxplot(data=bd_dados, x='Fechamento');
```

```python
#Analisar um boxplot Mensal para de uma forma ampla analisar o Fachamento
```

```python
dados['Mes'] = dados['Data'].dt.month

plt.figure(figsize=(16, 5))
sns.boxplot(data=dados, x='Mes', y='Fechamento', palette="Set1");
```

```python
#Demostra uma descrição atravez do describe de cada mês em relação ao fechamento?
```

```python
dados.groupby(['Mes']).describe()['Fechamento']
```
